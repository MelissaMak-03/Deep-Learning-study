```python
from PIL import Image
from pathlib import Path
import os

def resize_images(image_path, max_size=400, dest=None):
    for img_file in Path(image_path).iterdir():
        if img_file.suffix.lower() == '.heic':
            print(f"Skipping .heic file: {img_file}")
            continue
        
        try:
            with Image.open(img_file) as img:
                img.thumbnail((max_size, max_size))
                dest_path = dest / img_file.name if dest else img_file
                img.save(dest_path)
        except Exception as e:
            print(f"Error processing {img_file}: {e}")

# for path in searches[0]:
for country in searches["America"]:
    curr_path = Path("America")
    dest = (curr_path/country)
    dest.mkdir(exist_ok=True, parents=True)
    print(dest)
    for numb in notes:
        curr = f'{numb} {searches["America"][country]}'
        #     for curr in amer_searches[country]:
        folder = (dest/curr)
        print(folder)
        download_images(folder, urls=search_images(f'{curr} photo'))
        sleep(10)
        resize_images(dest/curr, max_size=400, dest=dest/curr)
```

```python
dls = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=[Resize(192, method='squish')]
).dataloaders(Path('America'), bs=32)

dls.show_batch(max_n=12)
```

```python
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(3)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```
